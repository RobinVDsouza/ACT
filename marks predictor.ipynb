```python
x='n'
count=0
pos=[]
for i in range(train_data.shape[0]):
    if train_data['external_marks'][i]>train_data['external_max'][i]:
        x='y'
        count+=1
        pos.append(i)
    if train_data['internal_marks'][i]>train_data['ia_max'][i]:
        x='y'
        count+=1
        pos.append(i)
print(x)
print(count)
print(pos)
```

```python
y=train_data.loc[:,['total_marks']]
```

```python
y.head()
```

```python
var='academic_period'
period=train_data[[var]]
period=pd.get_dummies(period)
period
```

```python
var='programme'
programme=train_data[[var]]
programme=pd.get_dummies(programme)
programme
```

```python
var='subject_name'
sub=train_data[[var]]
sub=pd.get_dummies(sub)
sub
```

```python
final_test=pd.concat([train_data,programme,period,sub],axis=1)
final_test.head()
```

```python
final_test.drop(["programme"],axis=1,inplace=True)
final_test.drop(["academic_period"],axis=1,inplace=True)
final_test.drop(["subject_name"],axis=1,inplace=True)
final_test.head()
```

```python
for i in final_test.columns:
    print(i)
```

```python
x=final_test.loc[:,['ia_max', 'internal_marks','external_max','programme_B.A.',
'programme_B.B.A.',
'programme_B.B.M.',
'programme_B.C.A.',
'programme_B.Com.',
'programme_B.S.W.',
'programme_B.Sc.',
'programme_M.A.',
'programme_M.Com.',
'programme_M.Sc.',
'academic_period_SEMESTER I',
'academic_period_SEMESTER II',
'academic_period_SEMESTER III',
'academic_period_SEMESTER IV',
'academic_period_SEMESTER V',
'academic_period_SEMESTER VI',]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=25)
```

```python
from sklearn.linear_model import LinearRegression
linear_reg=LinearRegression()
linear_reg.fit(X_train,Y_train)
y_pred=linear_reg.predict(X_test)
print("Accuracy on training data : ",linear_reg.score(X_train,Y_train))
print("Accuracy on test data : ",linear_reg.score(X_test,y_pred))
```

```python
from sklearn import linear_model
lasso=linear_model.Lasso(alpha=0.1)
lasso.fit(X_train,Y_train)
y_pred=lasso
lasso.predict(X_test)
print("Accuracy on training data: ",lasso.score(X_train,Y_train))
print("Accuracy on test data: ",lasso.score(X_test,Y_test))
```

```python
import pickle
pickle.dump(linear_reg,open('model1.pk1','wb'))
model=pickle.load(open('model1.pk1','rb'))
```

```python
def pred_totalmarks(ia_max,int_marks,ext_max,prog,period):
    prog='programme_'+prog
    prog_index=np.where(X_test.columns==prog)[0][0]
    period='academic_period_'+period
    period_index=np.where(X_test.columns==period)[0][0]
    
    
    x=np.zeros(len(X_test.columns))
    x[0]=ia_max
    x[1]=int_marks
    x[2]=ext_max
    if prog_index>=0:
        x[prog_index]=1
    if  period_index>=0:
        x[period_index]=1
        
    return model.predict([x])[0][0]
```
